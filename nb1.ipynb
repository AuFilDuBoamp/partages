```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity ='all'
from IPython.display import clear_output, display

import pandas as pd
pd.set_option('display.max_colwidth', 200)
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 200)
```

```python
df0 = pd.read_parquet('../0000_PREPA_BOAMP_NOTNA_2024_POUR_PATCH/EXPORT_AFDB_MERGE_PREPARE/boamp_siren_acheteurs_tout_2024.parquet')
```

```python
# catégoriser
df1 =df0.copy()
# catégories
for x in df1.columns:
    try:
        if df1[f'{x}'].nunique() <= 50:
            print(f'{x} : ' + str(df1[f'{x}'].nunique()) + '-> category.')
            df1[f'{x}'] = df1[f'{x}'].astype('category')
        else:
            df1[f'{x}'] = df1[f'{x}'].astype('string')
            print(f'{x} : ' + str(df1[f'{x}'].nunique()) + '-> string.')
            continue
             
    except Exception as e:
        print(e)
    finally:
        pass
 
df1.shape     
df1.info()           
```

```python
dfparis = (df1
           .loc[df1.SN_30_siren.eq('217500016'),:]
           .sort_values(by='B_09_dateParution', ascending=False)
           .reset_index(drop=True)
           .copy()
          )
```
